```python
# power=pd.DataFrame(df.iloc[:,5])#定位“对象”所在的列
# power#输出检查  
# power_count=power.groupby('所属势力').size()#对这一列中相同的元素计数
# power_count=pd.DataFrame(power_count)#放到dataframe表格里面
# power_count.columns=['频数']#把列名自定义一个名字，方便后续操作
# power_count_grouped=power_count.sort_values(by='频数',ascending=False)#按频数大小降序排序
# temp = 10#定义一个界限
# than_ten = power_count_grouped[power_count_grouped['频数']>=temp]#大于temp的数据保留
# than_ten
# lower_ten = power_count_grouped[power_count_grouped['频数'] < temp]['频数'].sum()#统计小于temp的数据数据
# other_row = pd.DataFrame([{'频数': lower_ten,'所属势力': '其他'}])#把所有小于temp的数据归类到“其他”里面（自己定义）
# than_ten['所属势力'] = than_ten.index#把索引值赋值给一个新的列
# # than_ten
# than_ten.reset_index(drop=True, inplace=True)#解决{power_count=power.groupby('所属势力').size()}这串代码导致的索引缺行问题,使其能对齐
# than_ten
# result=pd.concat([than_ten,other_row])#拼接两个表格
# explode = [0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02]#设置各个分块的间距
# colors=['lavender','g','r','c','m','y','cyan','pink','orange','grey','brown']
# plt.figure(figsize=(15,15))
# plt.pie(result['频数'],explode=explode,labels=result['所属势力'],colors=colors,autopct='%2.1f%%',textprops={'fontsize': 10})
# plt.show()
# plt.savefig('所属势力.png')
# result=pd.concat([than_ten,lower_ten],ignore_index=False)
power_count = df['所属势力'].value_counts().reset_index()
# print(power_count)
power_count.columns = ['所属势力', '频数']
temp = 10  # 设定频数界限
than_ten = power_count[power_count['频数'] >= temp].copy()
lower_than_ten_sum = power_count[power_count['频数'] < temp]['频数'].sum()

# 将小于频数界限的数据合并到一个 '其他' 类别
other_row = pd.DataFrame([{'所属势力': '其他', '频数': lower_than_ten_sum}])
result=pd.concat([than_ten,other_row])
# 可选的饼图配置
explode = [0.02] * len(result)  # 各个分块的间距
colors = ['lavender', 'g', 'r', 'c', 'm', 'y', 'cyan', 'pink', 'orange', 'grey'][:len(than_ten)]

# 绘制饼图
plt.figure(figsize=(10, 10))
plt.pie(result['频数'], explode=explode, labels=result['所属势力'],
        colors=colors, autopct='%2.1f%%', textprops={'fontsize': 10})
plt.title('所属势力分布')
plt.savefig('所属势力.png')
plt.show()
```
